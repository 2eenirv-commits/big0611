# 막대그래프(barplot)
- 카테고리별 수치 비교
- 데이터 포인트가 적은 경우(20개 미만)
-정확한 수치 값을 강조하고 싶을 때
- 순위나 크기 순서를 명확히 보여주고 싶을 때

```python
# tips 데이터를 활용한 막대그래프 생성
import seaborn as sns
df= sns.load_dataset('tips')
```

```python
# 1. 기본 막대그래프
sns.barplot(x=df['size'], y=df['tip'])
```

```python
# 1. 가로 막대그래프
# orient: 막대의 방향(v: 세로(기본값), h: 가로)
sns.barplot(x=df['tip'], y=df['size'], orient= 'h')
```

```python
# 2. 신뢰구간 제거
sns.barplot(x=df['size'], y=df['tip'], errorbar=None)
```

```python
# 3. 요일별 팁
sns.barplot(x=df['day'], y=df['tip'])
```

## 개수 세기 전용 막대그래프
### sns.countplot(x): y축 값은 x축 변수의 개수(count)로 표시

```python
sns.countplot(x=df['day'])
```

```python
df['day'].value_counts()
```

## 막대그래프의 변형
### sns.pointplot(x, y): 막대를 제외하고, 데이터의 평균값을 선으로 잇고, 신뢰구간을 함께 표시한 그래프

```python
sns.pointplot(x=df['day'], y=df['tip'])
```
