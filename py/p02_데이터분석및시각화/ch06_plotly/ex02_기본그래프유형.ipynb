# 기본 그래프 유형
## 선 그래프(line plot)

```python
import plotly.express as px
import pandas as pd
import numpy as np
np.random.randn(30)
```

```python
# 시계열 데이터 생성
# 2024-01-01 부터 D(Daily) 하루씩 30일간
dates = pd.date_range('2024-01-01', periods=30, freq='D')
# np.random.randn(30): 평균 0, 표준편차 1에 따른 난수 30개
values = np.cumsum(np.random.randn(30)) + 100

time_series = pd.DataFrame({
    '날짜': dates,
    '주가': values
})
time_series
```

```python
# 1. 단일 선 그래프
fig = px.line(time_series, x='날짜', y='주가',
              title='일별 주가 변동',
              markers=True)
fig.show()
```

```python
# 2. 다중 선 그래프
# 데이터 생성
companies = ['삼성전자', 'LG전자', 'SK하이닉스']
stock_data = []

for company in companies:
    for i, date in enumerate(dates):
        price = 100 + np.cumsum(np.random.randn(1))[0] * (i + 1) /10
        stock_data.append({'날짜': date, '회사': company, '주가': price})

# 데이터 분석
df_stocks = pd.DataFrame(stock_data)
df_stocks
```

```python
# 데이터 시각화
fig = px.line(df_stocks, 
              x = '날짜', # x축 칼럼명
              y = '주가', # y축 칼럼명
              color='회사', 
              title='회사별 주가 비교',
              line_dash='회사')
fig.show()
```

# 산점도(Scatter plot)

```python
# 1. 기본 산점도
# 키와 몸무게 데이터
height_weight = {
    '키': [160, 165, 170, 175, 180, 162, 168, 172, 178, 185],
    '몸무게': [55, 60, 65, 70, 75, 58, 63, 68, 73, 80],
    '성별': ['여성', '여성', '여성', '남성', '남성', 
           '여성', '여성', '남성', '남성', '남성']
}

df_body = pd.DataFrame(height_weight)
df_body

```

```python
fig = px.scatter(df_body, x='키', y='몸무게',
                     title='키와 몸무게의 관계')
fig.show()
```

```python
# 2. 그룹별 산점도
# 성별로 구분한 산점도
fig = px.scatter(df_body, # 데이터 프레임
                x='키', # x축 컬럼 -> x축 값 및 이름
                y='몸무게', # y축 컬럼 -> y축 값 및 이름
                color='성별', # 그룹별 색상 -> 범례
                title='성별에 따른 키와 몸무게 관계', # 그래프 제목
                size_max=15 # 마커의 최대 크기
                )

# 축 제목 변경
fig.update_layout(
    xaxis_title='키 (cm)',
    yaxis_title = '몸무게 (kg)'
)

fig.show()
```

```python
# 3. 버블 차트
# 크기까지 고려한 버블 차트
# 데이터 생성(직접, 크롤링)
countries = {
    '국가': ['한국', '일본', '중국', '미국', '독일'],
    'GDP': [1800, 4200, 14000, 21000, 3800],
    '인구': [51, 126, 1400, 330, 83],
    '대륙': ['아시아', '아시아', '아시아', '북미', '유럽']
}

# 데이터 분석(전처리 -> 데이터 다듬기, 통계)
df_countries = pd.DataFrame(countries)
df_countries
```

```python
# 머신러닝(기계 학습)
# 데이터 시각화
fig = px.scatter(df_countries, 
                 x ='GDP',
                 y='인구',
                 size='GDP',
                 color='대륙',
                 hover_name='국가',
                 title='국가별 GDP와 인구 관계',
                 size_max=60,
                 labels={ # 축 제목
                     'GDP': 'GDP (억 달러)',
                     '인구': '인구 (백만명)'
                 }
                )

# fig.update_layout(
#     xaxis_title='GDP (억 달러)',
#     yaxis_title='인구 (백만명)'
# )

fig.show()
```

## 히스토그램(Histogram)

```python
# 1. 기본 히스토그램
# 시험 점수 분포 데이터
# 데이터 생성
np.random.seed(42)  # 42 숫자는 의미x 관례적 숫자, seed는 난수 고정 역할
# 평균 75, 표준편차 15
scores = np.random.normal(75, 15,1000) # (평균, 표준편차, 랜덤숫자)
# 0~100 사이로 제한
scores = np.clip(scores, 0, 100) # (배열, 최소값, 최대값)

# len(scores) # 갯수
scores
```

```python
# 데이터 분석
df_hist = pd.DataFrame({'점수': scores})
df_hist
```
