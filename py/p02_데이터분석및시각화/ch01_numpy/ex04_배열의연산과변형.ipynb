```python
# 배열의 연산과 변형
## 벡터화 연산(Vectorized Operations)
import numpy as np

# 배열 생성
arr = np.arange(1,5)
arr
```

```python
## 벡터화 연산
arr + 2 
```

```python
arr * 2
```

```python
# 리스트
list = [1, 2, 3, 4]
list
```

```python
list + 2
```

```python
# 반복
list * 2
```

```python
arr = np.arange(1,5)
print(arr)
arr1 = np.arange(5, 9)
print(arr1)
```

```python
arr + arr1
```

```python
arr * arr1
```

```python
list = [1, 2, 3, 4]
print(list)
list1 = [5, 6, 7, 8]
print(list + list1)
```
