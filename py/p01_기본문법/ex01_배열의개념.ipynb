```python
# 외부 라이브러리는 설치 후 import해야 한다.
import numpy as np
```

```python
# 리스트의 항목에 2를 곱하기
# 예) [1, 2, 3, 4] -> [2, 4, 6, 8]
# 리스트 연산 - 반복문 필요!
list_data = [1, 2, 3, 4]
result_list = []
for item in list_data:
    result_list.append(item * 2)
result_list
print(result_list)    
```

```python
# 배열 연산 -> 반복문 불필요!
# 벡터화 연산
array_data = np.array([1, 2, 3, 4])
result_array = array_data * 2
result_array
print(result_array)
```

```python
# np.array() 함수   /np는 모듈 -> 모듈.함수()
list1 = [1, 2, 3]
arr1 = np.array(list1)
arr1
```

```python
# 리스트 vs 배열 성능 비교
import time
```

```python
# 리스트를 사용한 연산
start_time = time.time()
# 1970-01-01 00:00:00 기준으로 초 단위 시간을 표현한 것
print(start_time)
# [0, 1, 2, ..., 999999] 형태의 리스트 생성
python_list = list(range(1000000))
result_list = [x * 2 for x in python_list]
#print(result_list) 출력을 못함 
list_time = time.time() - start_time
print(f"리스트 연산 시간: {list_time:.4f}초")
```

```python
# NumPy 배열을 사용한 연산
start_time = time.time()
print(start_time)
numpy_array = np.arange(1000000)
result_array = numpy_array * 2
print(result_array) # 추가내용 출력가능
numpy_time = time.time() - start_time
print(f"NumPy 연산 시간: {numpy_time:.4f}초")
print(f'속도 개선: {list_time / numpy_time:.1f}배')
```
