```python
import numpy as np
```

```python
# 배열의 인덱싱과 슬라이싱
```

```python
## 1차원 배열의 인덱싱
```

```python
arr = np.arange(16)
print(arr)
```

# 1. 기본 인덱싱

```python
# 기본 인덱스
# 배열명[인덱스]
print(arr[5])
```

```python
# 마이너스 인덱스
print(arr[-1])
```

```python
# 슬라이싱: 인덱스 범위
print(arr[5:8])
```

```python
# 하나의 값 할당 - 배열 값 수정
arr[5] = 100
print(arr)
```

```python
# 여러 값 할당
arr[6:8] = [200,201]
print(arr)
```

```python
## 다차원 배열의 인덱싱
```

```python
# 다차원 배열 생성 (메서드 체이닝)
arr2 = np.arange(16).reshape(4,4)
print(arr2) 
```

```python
### 1. 2차원 배열의 인덱싱
- 배열명[행인덱스][열인덱스]
- 배열명[행인덱스, 열인덱스]
```

```python
# 한 행 전체 선택
print(arr2[1])
```

```python
# 개별 요소 선택 -> 두 가지 방법이 있다.
# 1행 1열
print(arr2[1][1])
print(arr2[1, 1])
```

```python
# 값 할당
arr2[1,2] = 100
print(arr2)
```

```python
### 2차원 배열 슬라이싱
- 배열명[s:e]
- 배열명[s:e, s:e]
```

```python
# 행 슬라이싱
print(arr2[0:2])
print(arr2[:2])
```

```python
### 3. 2차원 슬라이싱(행과 열 동시 지정)
```

```python
# 특정 행과 열 범위 선택
print(arr2[0:2, 2:4])
print(arr2[:2, 2:4])
```
